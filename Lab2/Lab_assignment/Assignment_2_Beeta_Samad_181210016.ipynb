### Task 3: 1. On the dataframe generated above, please answer the following:
<br>1. Extract Data of only India and make into a new dataframe. 
<br>2. Extract only the total cases column. 
<br>3. Convert total cases column into percentage. Percentage of total cases. Total cases is the number of cases as on the last date of the dataset. 
#### Use Numpy. DO NOT WRITE A LOOP! 
<br>4. Add this newly generated column to the India Dataframe.

```python
#1. Extracting data of only India:

#Extracting only the indian rows:
indiaDataFrame = dataFrame.loc[dataFrame['location']=='India']
#Resetting the indices:
indiaDataFrame = indiaDataFrame.reset_index() 
indiaDataFrame = indiaDataFrame.drop(columns=['index'])

print(indiaDataFrame)
```

```python
#2. Extract only the total cases column:

totalCasesDf = indiaDataFrame['total_cases']
print(totalCasesDf)
```

```python
#3. Convert total cases to percentage:
#importing numpy:
import numpy as np

lastDayCases = list(indiaDataFrame.sort_values(['date']).tail(1).total_cases)[0]
percentageDf = np.array(indiaDataFrame['total_cases'])/lastDayCases*100

print(percentageDf)
```

```python
#4. Add to the indian dataframe:

indiaDataFrame['percentage'] = percentageDf

print(indiaDataFrame)
```

### Task 5: 1. On the dataframe generated above, please answer the following:
<br>1. Extract data only for 18-08-2020. 
<br>2. which country has the max and min 
<br>3. take the top 10 and bottom 10 of total cases and total deaths

```python
#1. Extract only for 18-08-2020

dataFrame.loc[dataFrame['date']=='2020-08-18']
```

```python
#2. Which country has the max and min number of total_cases (assuming this is the question)

#extracting a dataframe without the location as 'world':
countryDataFrame = dataFrame.loc[dataFrame['location']!='World']
#extracting the dataframe with the latest date:
countryDataFrame = countryDataFrame.loc[countryDataFrame['date']==countryDataFrame.max()['date']]

maxCases = countryDataFrame.max()['total_cases']
minCases = countryDataFrame.min()['total_cases']

#finding max and min total_cases countries:
maxCountry = countryDataFrame.loc[countryDataFrame['total_cases']==maxCases]
minCountry = countryDataFrame.loc[countryDataFrame['total_cases']==minCases]

print("The country with the maximum number of cases: ", list(maxCountry.location)[0])
print("The country with the minimum number of cases: ", list(minCountry.location)[0])
```

```python
sortedTotalCases = dataFrame.sort_values(['total_cases'])
sortedTotalDeaths = dataFrame.sort_values(['total_deaths'])

print("Top 10 Total Cases: ", sortedTotalCases.head(10))
print("\nBottom 10 Total Cases: ", sortedTotalCases.tail(10))

print("\nTop 10 Total Deaths: ", sortedTotalDeaths.head(10))
print("\nBottom 10 Total Deaths: ", sortedTotalDeaths.tail(10))
```

```python
#1. b. total cases and death counts for each continent as of 18-08-2020

print("Total cases for each continent: \n", dataFrame.loc[dataFrame['date']=='2020-08-18'].groupby('continent').total_cases.sum())
print("\n\nTotal deaths for each continent: \n", dataFrame.loc[dataFrame['date']=='2020-08-18'].groupby('continent').total_deaths.sum())
```

```python
#2. Calculate mortality with apply()

def calculateMortality(row):
    return row[7]/row[4]*100

dataFrame['mortality_rate'] = dataFrame.apply(calculateMortality, axis=1)

dataFrame
```

```python
#3. Calculate mortality without apply() 

dataFrame['mortality_rate'] = dataFrame['total_deaths']/dataFrame['total_cases']*100

dataFrame
```
