```python
from bs4 import BeautifulSoup #pip install BeautifulSoup4
from urllib.request import urlopen, Request
```

```python
hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11,!(KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.,!8',
'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
'Accept-Encoding': 'none',
'Accept-Language': 'en-US,en;q=0.8',
'Connection': 'keep-alive'}
```

```python
webpage='https://www.cse.iitd.ac.in/index.php/2011-12-29-23-14-30/faculty'
```

```python
def extract_soup_object(link):
    request=Request(link, headers=hdr)
    html = urlopen(request).read().decode()
    soup = BeautifulSoup(html,'html.parser')
    return soup
```

```python
soup=extract_soup_object(webpage)
```

```python
soup.contents
```

```python
row_selector = '#centercontent_bg div div table tr'
rows = soup.select(row_selector)
```

```python
all_professors = []
```

```python
for r in rows:
    myDict = {}
    try:
        row = r.select('td:nth-child(2) > p')[0].getText().split("\n")
        emails = " ".join(r.select('td:nth-child(3) > p:nth-child(1)')[0].getText().strip().split())
        phones = " ".join(r.select('td:nth-child(3) > p:nth-child(2)')[0].getText().strip().split())
        specialisation = r.select('td:nth-child(2) > p:nth-child(2)')[0].getText()
        myDict['name'] = row[0].strip()
        myDict['designation'] = row[1].strip()
        myDict['specialisation'] = specialisation
        myDict['emails'] = emails
        myDict['phones'] = phones
        
        all_professors.append(myDict)
        
    except:
        print("Not a valid row")
```
